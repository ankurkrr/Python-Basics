### with open

Most common way to perform operations on files

it closes the file after performing operations

```python
with open('open.txt') as f:
    print(f.read())
```

```python
f.closed
```

```python
with open('open.txt','w') as f:
    f.write('hey iam writable')
```

### Writing in a file

You can write both text and binary file

You can wither write or append in a file

```python
# we r creating new file here
with open('open1.txt', "w") as f:
    f.write("hey new file open and created")
```

```python
# Write in existing file (overwrite)

with open('open1.txt', "w") as f:
    f.write("update 1 in this file")
```

```python
with open('open1.txt', "w") as f:
    f.write("update 1 in this file \n")
    f.write("update 2 in this file")
```

### Read

using read() method wr can read a file

it opens in read only format

```python
with open("open1.txt",'r') as f:
    data = f.read()
    print(data)
```

read() : reads the whole data

read(): reads data of length
    
tell(): tells about the position of the file handle
    
seek(): it helps to reposition your file handle
    
readline() : reads data line by line

```python
with open("open1.txt",'r') as f:
    
    # Read first 4 characters
    data = f.read(4)
    print(data)
    
    print(f.tell())
    
    #next 10 character
    data = f.read(10)
    print(data)
    
    # tell: tells you the position of the file handle
    print(f.tell())
    
    # using seek you can reset your file handler
    f.seek(3)
    
    print(f.tell())
    print(f.read())
```

```python
with open("open1.txt",'r') as f:
    data = f.readlines()
    print(data)
```

### BINARY  DATA
To read binary data we open file in "rt" mode

To write binary data we open file in "wt" mode


```python
with open('Photo.jpg', "rb") as f:
    data = f.read()
    print(data)
```

```python
with open('Photo.jpg', "rb") as f:
    data = f.read()
    
    with open('new.jpg', 'wb') as d:
        d.write(data)
```
